```python
rec0= pd.read_csv("nexo_transactions.csv")
```

```python
df0= rec0.rename(columns={"Outstanding Loan":"Loan", "Date / Time":"Date"}) #rename column name
```

```python
df0["Date"] = pd.to_datetime(df0["Date"]) #convert date time
```

```python
df0.replace({'NEXONEXO':'NEXO'}, regex = True, inplace = True)  #replace NEXO
```

```python
df0.replace({'NEXOBEP2':'NEXO'}, regex = True, inplace = True)  #replace NEXOBEP2
```

```python
df0.replace({'NEXOUSDTERC':'NEXO'}, regex = True, inplace = True)  #replace NEXOUSDTERC
```

```python
df0.replace({'\$':''}, regex = True, inplace = True)  #replace $
```

```python
df0['Loan'] = df0['Loan'].astype('float64')
```
